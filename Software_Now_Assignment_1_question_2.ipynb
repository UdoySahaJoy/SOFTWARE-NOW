```python
#Assignment question 2
number_limit= int(input("Input: "))
total_prime_number = 0
total_sum = 0
smallest_prime_number = 0
largest_prime_number = 0
print("Output:")
print("Prime numbers found:")
for num in range(2, number_limit+1):
    prime_number = True
    i = 2
    while i<num:
        if num % i == 0:
            prime_number = False
            break
        i=i+1
    if prime_number == True:
        print(num,end=" ")
        total_prime_number = total_prime_number+1
        total_sum = total_sum + num

        if total_prime_number ==1:
            smallest_prime_number = num
            largest_prime_number = num
        else:
            if num<smallest_prime_number:
                smallest_prime_number = num
            if num>largest_prime_number:
                largest_prime_number = num


print("\nTotal primes found:",total_prime_number)

if total_prime_number>0:
    print("Largest prime: ",largest_prime_number)
    print("Smallest prime: ",smallest_prime_number)
    print("Sum of all primes: ",total_sum)
else:
    print("No primes found")
```
