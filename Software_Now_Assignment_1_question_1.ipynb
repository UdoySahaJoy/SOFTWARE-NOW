```python
#Assignment question 1
user_password = input("Enter your password: ")
pass_length = 0
digit = False
upper_case = False

for i in user_password:
    pass_length= pass_length+1

    if i>='0' and i<='9':
        digit = True
    if i >= "A" and i<= "Z":
        upper_case = True


if pass_length<6:
    print("Weak password")
elif pass_length<=10:
    if digit == True:
        print("Medium password")
    else:
        print("Weak password")
else:
    if digit == True and upper_case == True:
        print("Strong password")
    else:
        print("Medium password")
```
